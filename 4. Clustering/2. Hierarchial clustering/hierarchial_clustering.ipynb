```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
dataset = pd.read_csv('Mall_Customers.csv')
x = dataset.iloc[:, 3:].values
# we are taking only two feature here..so that we can visualize it
```

No need to split in training and test data. Because we don't have dependent variable here.

```
# dendro gram
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(x,method='ward'))
plt.title('Dendrogram')
plt.xlabel('Data Points[customers]')
plt.ylabel('Euclidean distance')
plt.show()
```

```
# model training
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
y_hc = hc.fit_predict(x)
print(y_hc)
```

```
# visualising
plt.scatter(x[y_hc==0,0],x[y_hc==0,1],s=100,color='red', label = 'Cluster 1')
plt.scatter(x[y_hc==1,0],x[y_hc==1,1],s=100,color='black', label = 'Cluster 2')
plt.scatter(x[y_hc==2,0],x[y_hc==2,1],s=100,color='green', label = 'Cluster 3')
plt.scatter(x[y_hc==3,0],x[y_hc==3,1],s=100,color='blue', label = 'Cluster 4')
plt.scatter(x[y_hc==4,0],x[y_hc==4,1],s=100,color='magenta', label = 'Cluster 5')
plt.title('Customers clustering')
plt.xlabel('Salary')
plt.ylabel('Spending score')
plt.legend()
plt.show()
```
